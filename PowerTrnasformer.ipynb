```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from scipy.stats import stats
```

```python
df=pd.read_csv('concrete_data.csv')
df.head()
```

```python
x=df.drop(columns=['Strength'])
y=df['Strength']
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

#### Box-Cox transformation

```python
trf1 = PowerTransformer(method='box-cox')
x_train_box  = trf1.fit_transform(x_train+0.000001)
x_test_box  = trf1.transform(x_test+0.0000001)
x_train_box=pd.DataFrame(x_train_box)
x_test_box=pd.DataFrame(x_test_box)
```

```python
LR = LinearRegression()
LR.fit(x_train_box,y_train)
y_pred=LR.predict(x_test_box)
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(LR,x_train_box, y_train, cv=5, )
score.mean()
```
